```python
def ovrMetrics(C):
    N = C.shape[0]
    total = np.sum(C)
    
    lst = {
        "P": np.zeros(N, dtype=np.uint32),
        "N": np.zeros(N, dtype=np.uint32),
        "TP": np.zeros(N, dtype=np.uint32),
        "FP": np.zeros(N, dtype=np.uint32),
        "FN": np.zeros(N, dtype=np.uint32),
        "TN": np.zeros(N, dtype=np.uint32),
        "CP": np.zeros(N, dtype=np.uint32),
        "CN": np.zeros(N, dtype=np.uint32),
        "TPR": np.ones(N),
        "TNR": np.ones(N),
        "PPV": np.zeros(N),
        "NPV": np.zeros(N),
        "FPR": np.zeros(N),
        "FNR": np.zeros(N),
        "FDR": np.zeros(N),
        "FOR": np.zeros(N),
        "PT": np.zeros(N),
        "ACC": np.zeros(N),
    }
    
    # Let $y$ denote the ground truth label
    # True Count: $P_c = |\{y_i = c}|$ for all categories $c\in C$.
    lst["P"] = np.sum(C, axis=1) # stores count
    # True Negative: $N_c = \{y_i \neq c}$ for all categories $c\in C$.
    lst["N"] = total - lst["P"] # stores count
    
    # Let $\tilde{\cdot}$ denote a prediction
    # Predicted Positive: $\tilde{P}_c = \{\tilde{y}}_i = c}$ for all $c\in C$.
    lst["PP"] = np.sum(C, axis=0)
    # Predicted Positive: $\tilde{N}_c = \{\tilde{y}}_i \neq c}$ for all $c\in C$.
    lst["PN"] = total - lst["PP"]
    
    # True Positives: $TP_c = P_c \cap \tilde{P}_c$.
    lst["TP"] = np.diag(C)
    # False Negatives: $FN_c = P_c \cap \tilde{N}_c$.
    lst["FN"] = lst["P"] - lst["TP"]
    
    
    # False Positives: $FP_c = N_c \cap \tilde{P}_c$.
    lst["FP"] = lst["PP"] - lst["TP"]
    # True Negatives: $TN_c = N_c \cap \tilde{N}_c$.
    lst["TN"] = lst["N"] - lst["FP"]
    
    # True Positive Rate: $TPR_c = p(\tilde{y}_i = c|y_i = c) = \frac{|TP_c|}{|P_c|}$
    lst["TPR"] = np.divide(lst["TP"], lst["P"], where=lst["P"]!=0)
    
    # True Negative Rate: $TNR_c = p(\tilde{y}_i\neq c|y_i \neq c) = \frac{|TN_c|}{|N_c|}$
    lst["TNR"] = np.divide(lst["TN"], lst["N"], where=lst["N"]!=0)
    
    # False Negative Rate: $FPR_c = p(\tilde{y}_i\neq c|y_i = c) = \frac{|FN_c|}{|P_c|}$
    lst["FNR"] = np.divide(lst["FN"], lst["P"], where=lst["P"]!=0)
    
    # False Positive Rate: $FPR_c = p(\tilde{y}_i = c|y_i \neq c) = \frac{|FP_c|}{|N_c|}$
    lst["FPR"] = np.divide(lst["FP"], lst["N"], where=lst["N"]!=0)
    return lst
```
